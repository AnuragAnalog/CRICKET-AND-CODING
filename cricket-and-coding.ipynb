# CRICKET AND CODING

```python
# Required modules

import os
import glob
import numpy as np
import pandas as pd
import seaborn as sns

from zipfile import ZipFile
from matplotlib import pyplot as plt

from sklearn.preprocessing import LabelEncoder
```

```python
# Initial Configuration

%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 7)
```

```python
# Extracting the data

if not os.path.isfile('all_matches.csv'):
    with ZipFile('ipl_csv2.zip', 'r') as zf:
        zf.extractall('./')
```

```python
# Loading the data

data = pd.read_csv('all_matches.csv', index_col=['match_id'])
data.head()
```

```python
# Inspecting the data

data.info()
data.describe()
```

```python
# Defining the encoder's

venue_encoder = LabelEncoder()
batting_team_encoder = LabelEncoder()
bowling_team_encoder = LabelEncoder()
striker_encoder = LabelEncoder()
bowler_encoder = LabelEncoder()

venue_encoder.fit(data['venue'])
batting_team_encoder.fit(data['batting_team'])
bowling_team_encoder.fit(data['bowling_team'])
striker_encoder.fit(data['striker'])
bowler_encoder.fit(data['bowler'])
```

```python
# Collect data from each match

train = []
labels = []

for match in glob.glob('[0-9]*.csv'):
    match_data = pd.read_csv(match, index_col=['ball'])
    m = (match_data[['venue', 'innings', 'batting_team', 'bowling_team', 'striker', 'bowler']].iloc[0, :].tolist())
    runs = match_data[match_data['innings'] == 2].loc[0.1:4.5][['runs_off_bat', 'extras']].values.sum()
    train.append(m)
    labels.append(runs)
    
print(np.mean(labels))
```
